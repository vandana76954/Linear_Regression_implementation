```python
import pandas as pd
```

```python
#lets create dataframe
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
#Adding the target variable in dataframe
df["Price"]=dataset.target
```

```python
#1. Dividing the dataset into independent and dependent features
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Dividing the dataset into train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.33 ,random_state =42)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Feature Scaling --- Standardization-- with the help of z-score it will try to make mean=0 std=1
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
# Model_Training  Linear Regression Pocess 
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression() #initializing the object
```

```python
#For training the model we need to find out the values of thetas (coefficients)
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
#prediction w.r.t Linear Regression model
y_pred = regression.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error 
```

```python
import numpy as np
```

```python
#Errors
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
#Performance Metrics
#Accuracy R2 and Adjusted R2
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
```

```python
print(f"R-Square : {score}")
```

```python
#display adjusted R-squared
adjusted_score = 1 -(1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
print(f"Adjusted R-Square : {adjusted_score}")
```

```python
#Applying Ridge regression
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=20.0)
ridge.fit(X_train,y_train)
```

```python
#prediction with ridge
y_pred = ridge.predict(X_test)
```

```python
#Errors
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
#Applying Lasso regression
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=20.0)
```

```python
lasso.fit(X_train,y_train)
```

```python
#prediction with lasso
y_pred = lasso.predict(X_test)
```

```python
#Errors
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```
